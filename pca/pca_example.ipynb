```python
import matplotlib.pyplot as plt
import numpy as np
from pca import *

t = np.linspace(0, 2 * np.pi, 100)
alpha = 45
x=4*np.cos(t)
y=2*np.sin(t)

# Data
x = x *np.cos(alpha) + y* (-np.sin(alpha))
y = x *np.sin(alpha) + y* (-np.cos(alpha))

fig = plt.figure()
plt.plot(x,y, 'bo')
plt.show()
```

```python
# pca instance
X=np.vstack([x,y])
m=pca(X)
```

```python
# covariance matrix of the data
sigma = m.cov_matrix()
print(sigma)
```

```python
# eigenvectors and eigenvalues of the covariance matrix
D,V = m.eigs()
print(D)
print(V)
```

```python
fig = plt.figure()
plt.plot(x,y,'bo')

plt.arrow(0, 0, V[0,0], V[1,0], head_width=0.1, head_length=0.1, fc='r', ec='r')
plt.arrow(0, 0, V[0,1], V[1,1], head_width=0.1, head_length=0.1, fc='g', ec='g')
plt.show()
```

```python
# Project the data onto the eigenvectors to get the principal components
F=m.principal_comps()
```

```python
n=pca(F.T)
DD,VV = n.eigs()

fig = plt.figure()
plt.arrow(0, 0, VV[0,0], VV[1,0], head_width=0.1, head_length=0.1, fc='r', ec='r')
plt.arrow(0, 0, VV[0,1], VV[1,1], head_width=0.1, head_length=0.1, fc='g', ec='g')
plt.plot(F[:,0], F[:,1], 'bo')
plt.show()
```

```python
# Plot the Principal components
fig = plt.figure()
plt.plot(F[:,0], 'bo', label="first pc")
plt.plot(F[:,1], 'ro', label="second pc")
plt.title("Principal Components")
plt.legend()
plt.show()

fig = plt.figure()
plt.plot(x, 'go', label = "x")
plt.plot(y, 'yo', label = "y")
plt.title("Original Data")
plt.legend()
plt.show()
```

```python
# Fraction of variance explained by every component
v=m.var_explained()
print(v)
```
