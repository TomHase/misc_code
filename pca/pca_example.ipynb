```python
import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(0, 2 * np.pi, 100)
alpha = 45
x=4*np.cos(t)
y=2*np.sin(t)

x = x *np.cos(alpha) + y* (-np.sin(alpha))
y = x *np.sin(alpha) + y* (-np.cos(alpha))

fig = plt.figure()

plt.plot(x,y)
plt.show()
```

```python
sigma = np.cov(x,y)
```

```python
D,V = np.linalg.eig(sigma)

idx = D.argsort()[::-1]   
D = D[idx]
V = V[:,idx]
```

```python
fig = plt.figure()
plt.plot(x,y)

plt.arrow(0, 0, V[0,0], V[1,0], head_width=0.1, head_length=0.1, fc='r', ec='r')
plt.arrow(0, 0, V[0,1], V[1,1], head_width=0.1, head_length=0.1, fc='g', ec='g')
plt.show()
```

```python
W=V[:,:]
Z=np.stack((x.T,y.T), axis=1)
T=np.dot(Z,W)
```

```python
fig = plt.figure()
plt.plot(T[:,0],T[:,1])
plt.show()
```
