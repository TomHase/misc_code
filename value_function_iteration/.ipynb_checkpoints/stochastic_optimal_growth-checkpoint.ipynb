```python
import numpy as np
import matplotlib.pyplot as plt
from value_function_iteration import *

# parameters
alpha=0.33
beta=0.95
sigma = 2
delta = 0.1
mu=0
s=0.1

# tolerance
tolerance = 0.01

# capital steady state
kss = (alpha/((1/beta)-(1-delta)))**(1/(1-alpha))

# grid
grid_min = 0.25*kss
grid_max = 1.75*kss
grid_size = 100         
grid = np.linspace(grid_min, grid_max, grid_size)

# initial guess
v0 = np.zeros_like(grid)

# bellman equation
def bellman(kp, k, v_fn, alpha, beta, sigma, delta):
    c=k**alpha + (1-delta)*k - kp
    a1=1-sigma
    a2=beta*v_fn(kp)
    
    if c<=0:
        val = -888888888888888888-800*abs(c)
    else:
        val = (1/a1)*(c**a1-1) + a2
    
    return -val
              
    
policy, v = value_function_iteration(grid, tolerance, v0, alpha, beta, sigma, delta, bellman, grid_min, grid_max)
```

```python
# Value function
fig, ax = plt.subplots()
ax.plot(grid, v, label='value function')
ax.legend(loc='lower right')
ax.set_xlabel('k')
ax.set_ylabel('v(k)')
plt.show()

# Capital
fig, ax = plt.subplots()
ax.plot(grid, policy, label='policy function for capital')
ax.legend(loc='lower right')
ax.set_xlabel('k')
ax.set_ylabel("k'")
plt.show()

# Consumption
cons=grid**alpha + (1-delta)*grid - policy
fig, ax = plt.subplots()
ax.plot(grid, cons, label='policy function for consumption')
ax.legend(loc='lower right')
ax.set_xlabel('k')
ax.set_ylabel("c")
plt.show()
```
